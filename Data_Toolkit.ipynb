 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a sample dataset
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],
    'C': [5, 3, 8, 6, 7],
    'D': [10, 15, 20, 25, 30]
}

# Converting to DataFrame
df = pd.DataFrame(data)

# Calculating the correlation matrix
corr_matrix = df.corr()

# Plotting the heatmap
plt.figure(figsize=(6, 4))  # Set figure size
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Display the plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Create a DataFrame and add a new column based on an existing column

```python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
}

df = pd.DataFrame(data)

# Adding a new column based on the existing 'Age' column
df['Age Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Adult')

# Display the updated DataFrame
print(df)
```

Create a line plot with multiple lines using MatplotlibA

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]  # X-axis values
y1 = [2, 4, 6, 8, 10]  # First line
y2 = [1, 3, 5, 7, 9]   # Second line
y3 = [2, 3, 5, 6, 8]   # Third line

# Plotting multiple lines
plt.plot(x, y1, marker='o', linestyle='-', color='b', label='Line 1')
plt.plot(x, y2, marker='s', linestyle='--', color='r', label='Line 2')
plt.plot(x, y3, marker='^', linestyle='-.', color='g', label='Line 3')

# Adding labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Multiple Lines")

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

<<A Perform matrix multiplication using NumPyA

```python
import numpy as np

# Defining two matrices
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Performing matrix multiplication
result = np.dot(A, B)  # Method 1
# result = A @ B      # Method 2 (alternative)

# Displaying the result
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", result)
```
